Step 3: droping out non value added columns

```python
# droping some duplicate or non value added columns
df_store.drop(columns=['Customer ID','Postal Code','Country', 'Product ID'], inplace=True, axis=1)
df_store
```

Step 4: Aligning the date and time format 

```python
#infer_datetime_format : boolean, default False
#If True and parse_dates is enabled, pandas will attempt to infer the format of the datetime strings in the columns, and if it can be inferred, switch to a faster method of parsing them. In some cases this can increase the parsing speed by 5-10x.
#Essentially, Pandas deduces the format of your datetime from the first element(s) and then assumes all other elements in the series will use the same format. This means Pandas does not need to check multiple formats when attempting to convert a string to datetime.
df_store['Ship Date'] = pd.to_datetime(df_store['Ship Date'], infer_datetime_format=True)
df_store['Order Date'] = pd.to_datetime(df_store['Order Date'], infer_datetime_format=True)
print(df_store['Ship Date'],df_store['Order Date'])
```

Step 5: Manipulating the existing variable to get some more insights 

```python
# 5.1 - Arriving 4 new columns based on the existing continuous variables
df_store['unit_sales_price'] = df_store['Sales']/df_store['Quantity'] 
df_store['unit_quotation_price'] = (df_store['Sales']/df_store['Quantity'])/(1-df_store['Discount'])
df_store['unit_product_cost'] = (df_store['Sales']-df_store['Profit'])/df_store['Quantity']
df_store['profit%'] = ((df_store['unit_sales_price']-df_store['unit_product_cost'])/ df_store['unit_sales_price'])*100
df_store
```

```python
# 5.2 - Extracting years, months, weekdays from the datestamp for analysis.
df_store['order_year'] = df_store['Order Date'].dt.year
# where 2014 indicates year 2014 and so on...
df_store['month_of_order_year'] = df_store['Order Date'].dt.month
# where 1-Jan, 2-Feb, 3-Mar,.....,12-Dec.
df_store['weekday_of_order_year'] = df_store['Order Date'].dt.weekday 
# where MONDAY is 0 and SUNDAY is 6
df_store.head()
```

Step 7: Data Preparation for modeling..

```python
#Extrating the following attributes from the given data, which we feel that it may have an impact on the target attribute
#7.1 - 'quarters1-4' - dividing the years into 4 quarters.

def func(x):
    if x==1 or x==2 or x==3:
        return "1"
    if x==4 or x==5 or x==6:
        return "2"
    if x==7 or x==8 or x==9:
        return "3"
    if x==10 or x==11 or x==12:
        return "4"
    
df_store['quarters1-4'] = df_store['month_of_order_year'].apply(lambda x : func(x))
df_store['quarters1-4'] = df_store['quarters1-4'].astype('int32')
```

```python
#7.2 - 'Order ID_cat' - one order has multiple line items, does that orders with high no. of line items and low no. of line items have an impact on target attribute.

dictT=df_store['Order ID'].value_counts()
dict1=dictT.to_dict()
df_store['Order ID_cat']=df_store['Order ID'].map(dict1)
```

```python
#7.3 - label encoding - converting all the useful categorical data into numerical data.

#pull all the Cat variable into separate dataset
obj_df = df_store.select_dtypes(include=['object']).copy() 

# check for null values
obj_df[obj_df.isnull().any(axis=1)] 

#convert the data type from object to category
obj_df [['Ship Mode', 'Customer Name','Segment','City','State','Region','Category','Sub-Category','Product Name']] = obj_df[['Ship Mode', 'Customer Name','Segment','City','State','Region','Category','Sub-Category','Product Name']].astype('category')

# encode to integers
obj_df['Ship Mode_cat'] = obj_df['Ship Mode'].cat.codes
obj_df['Customer Name_cat'] = obj_df['Customer Name'].cat.codes
obj_df['Segment_cat'] = obj_df['Segment'].cat.codes
obj_df['City_cat'] = obj_df['City'].cat.codes
obj_df['State_cat'] = obj_df['State'].cat.codes
obj_df['Region_cat'] = obj_df['Region'].cat.codes
obj_df['Category_cat'] = obj_df['Category'].cat.codes
obj_df['Sub-Category_cat'] = obj_df['Sub-Category'].cat.codes
obj_df['Product Name_cat'] = obj_df['Product Name'].cat.codes

#merge the new dummy dataset (obj_df) with the parent dataset (df_store).
mod_df = pd.merge(df_store, obj_df,how="inner",
    on=None,
    left_on=None,
    right_on=None,
    left_index=False,
    right_index=False,
    sort=True,
    suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None,
)

#check the newly modified dataset mod_df
#mod_df.columns.values.tolist()
#mod_df.info()
```

Step 9: Linear Regression

```python
# for sales vs profit:

from sklearn.linear_model import LinearRegression
model = LinearRegression()

plt.scatter(mod_df['Sales'],mod_df['Profit'],c="#00000020")
plt.show() 

model.fit(mod_df[['Sales']],mod_df['Profit'])
print('intercept:', model.intercept_)
print('slope:', model.coef_)

r_sq = model.score(mod_df[['Sales']],mod_df['Profit'])
print('R2=',r_sq)

mod_df['pred'] = model.predict(mod_df[['Sales']])

plt.scatter(mod_df[['Sales']],mod_df['Profit'], c="#77000022")
plt.scatter(mod_df[['Sales']],mod_df['pred'],c="#00770022")
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()

mod_df['resid'] = mod_df['pred']-mod_df['Profit']
plt.boxplot(mod_df['resid'])
plt.show()
plt.hist(mod_df['resid'], bins=100)
plt.show()
mod_df['resid'].describe()
print('Mean of residual  {:.12f}'.format(mod_df['resid'].mean()))
print("Standard deviation of residual  {:.12f}".format(mod_df['resid'].std()))
print("Mean Profit {:.12f}".format(mod_df['Profit'].mean()))
print("Median Profit {:.12f}".format(mod_df['Profit'].median()))
```

Step 10: Multiple Linear Regression

```python
# considering the 2nd attribute lets see the results
model1 = LinearRegression()
model1.fit(mod_df[['Sales','Discount']],mod_df['Profit'])
print('intercept:', model1.intercept_)
print('slope:', model1.coef_)

r_sq = model1.score(mod_df[['Sales','Discount']],mod_df['Profit'])
print('R2=',r_sq)

mod_df['pred1'] = model1.predict(mod_df[['Sales','Discount']])

mod_df['resid1'] = mod_df['pred1']-mod_df['Profit']

plt.boxplot(mod_df['resid1'])
plt.show()
plt.hist(mod_df['resid1'], bins=100)
plt.show()
mod_df['resid1'].describe()
print('Mean of residual  {:.12f}'.format(mod_df['resid1'].mean()))
print("Standard deviation of residual {:.12f}".format(mod_df['resid1'].std()))
print("Mean Profit {:.12f}".format(mod_df['Profit'].mean()))
print("Median Profit {:.12f}".format(mod_df['Profit'].median()))
```

Step 11: K-Fold Cross Validation

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
import numpy as np

model = LinearRegression()

kf = KFold(n_splits=5, random_state=100, shuffle=True)

X = mod_df[['Sales','Discount']]
y = mod_df['Profit']

X = X.reset_index(drop=True)
y = y.reset_index(drop=True)

mse_values = []

for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_values.append(mse)
    
mean_mse = np.mean(mse_values)
print(mse_values, mean_mse)
```
